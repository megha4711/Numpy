**5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap**

```python
# Set random seed for reproducibility
np.random.seed(0)

# Generate synthetic data with correlated features
# Creating a random dataset with 100 samples and 3 features
n_samples = 100
feature_1 = np.random.normal(0, 1, n_samples)
feature_2 = 0.8 * feature_1 + np.random.normal(0, 0.5, n_samples)
feature_3 = 0.5 * feature_1 + 0.5 * feature_2 + np.random.normal(0, 0.5, n_samples)

# Creating a DataFrame
data = pd.DataFrame({'Feature 1': feature_1, 'Feature 2': feature_2, 'Feature 3': feature_3})

# Compute the correlation matrix
corr_matrix = data.corr()

# Creating a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5, fmt='.2f')

# Add title and labels
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.**

```python
from bokeh.plotting import figure, show

# Generate x values from 0 to 10
x = np.linspace(0, 10, 100)

# Generate y values as sine of x
y = np.sin(x)

# Create a new plot
plot = figure(title="Sine Wave", x_axis_label='x', y_axis_label='y', width=800, height=400)

# Add sine wave line to the plot
plot.line(x, y, legend_label="sin(x)", line_width=2)

show(plot)
```

**2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.**

```python
# Generating random data
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
sizes = np.random.randint(10, 50, size=n)  # Random sizes for the markers
colors = np.random.choice(['#f8c8d3', '#ff66b2', '#66ccff', '#ffcc00'], size=n)  # Random colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

# Create the scatter plot
p = figure(title="Bokeh Scatter Plot with Random Sizes and Colors",
           x_axis_label='X', y_axis_label='Y',
           tools="pan,box_zoom,reset")

p.scatter('x', 'y', size='sizes', color='colors', source=source, legend_field="colors", fill_alpha=0.6)

# Display the plot
output_notebook()
show(p)
```

**3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.**

```python
# Dataset
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the bar chart
p = figure(x_range=fruits, title="Fruit Counts",
           x_axis_label='Fruit', y_axis_label='Count',
           tools="pan,box_zoom,reset")

p.vbar(x='fruits', top='counts', width=0.6, source=source, legend_field="fruits", color="skyblue", alpha=0.7)

# Display the plot
output_notebook()
show(p)
```

**4. Create a Bokeh histogram to visualize the distribution of the given data**

```python
# Generate random data
data_hist = np.random.randn(1000)

# Calculate histogram
hist, edges = np.histogram(data_hist, bins=30)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:], bottom=np.zeros_like(hist)))

# Create the histogram plot
p = figure(title="Bokeh Histogram",
           x_axis_label='Value', y_axis_label='Frequency',
           tools="pan,box_zoom,reset")

p.quad(top='top', bottom='bottom', left='left', right='right', source=source,
       fill_color="skyblue", line_color="white", alpha=0.7)

# Display the plot
output_notebook()
show(p)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_file
from bokeh.models import ColorBar
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256

# Generate data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create Bokeh plot
output_file("heatmap.html")

# Flatten data for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
colors = data_heatmap.flatten()

# Create a colormap for coloring the heatmap
mapper = linear_cmap(field_name='color', palette=Viridis256, low=min(colors), high=max(colors))

# Create the figure
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', toolbar_location=None)
p.scatter(x_flat, y_flat, marker='square', size=15, color=mapper, line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], width=8, location=(0, 0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
